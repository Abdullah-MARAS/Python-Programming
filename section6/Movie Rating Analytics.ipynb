```python
sns.set_style('whitegrid')
list1 = []#list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre==gen].BudgetMillions)
#print(list1)
fig,ax=plt.subplots()#ceates one subplots
fig.set_size_inches(11.7,8.27)#size of A4 paper
h=plt.hist(list1,bins=30,stacked=True,rwidth=0.9,label=movies.Genre.cat.categories)
#we can also change of histogram by using color parameter. We should provide a list of colors
plt.title('Movie Budget Distribution',fontsize=35,\
         color='DarkBlue',fontname='Console')
plt.ylabel('Number of Movies',fontsize=25,color='Red')
plt.xlabel('Budget',fontsize=25,color='Green')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)#we can also change xticks values

plt.legend(frameon=True,fancybox=True,shadow=True,prop={'size':20},framealpha=1)
plt.show()
#if we don'tspecify the location of graph it goes to last place.
```
