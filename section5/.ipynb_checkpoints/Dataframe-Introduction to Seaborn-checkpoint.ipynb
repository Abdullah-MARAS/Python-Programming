```python
import seaborn as sns
plt.rcParams['figure.figsize']=8,4
```

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#Distribution:
vis1=sns.distplot(stats['InternetUsers'])
```

```python
vis1=sns.distplot(stats['InternetUsers'],bins=30)#shift+tab shows parameters
```

```python
#Boxplots:
vis2=sns.boxplot(data=stats,x='IncomeGroup',y='BirthRate')
```

```python
#Seaborn Galary
```

```python
sns.set()
vis3=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg=False)
```

```python
vis4=sns.distplot(stats['BirthRate'],bins=30)
```

```python
vis3=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',\
                fit_reg=False,hue='IncomeGroup',size=10,aspect=1)
```

```python
#marker size:
vis3=sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',\
                fit_reg=False,hue='IncomeGroup',size=10,aspect=1,\
                scatter_kws={'s':100})
```
